```python
X = music_data[['danceability', 'tempo']]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_sc = sc.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
k = 3

km = KMeans(n_clusters=k, random_state = 42)

km.fit(X_sc)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
pred = km.predict(X_sc)
```

```python
df['cluster'] = km.labels_
```

```python
df['predictions'] = pred
```

```python
df.head()
```

```python
centroids = pd.DataFrame(
sc.inverse_transform(km.cluster_centers_),
columns = ['danceability', 'tempo'])

centroids
```

```python
color_map = {0:"orange", 1:"blue", 2:"green"}
df['color'] = df['cluster'].map(color_map)

ax = df.plot(    
    kind="scatter", 
    x="danceability",
    y="tempo",
    figsize=(10,8),
    c = 'color'
)

centroids.plot(
    kind="scatter", 
    title = 'Clusters',
    x="danceability",
    y="tempo", 
    marker="*",
    c=centroids.index.map({0:"black", 1:"black", 2:"black"}),
    s=550,
    ax=ax
)
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_sc, km.labels_)
```

```python
scores = []

for k in range(2, 10):
    cl = KMeans(n_clusters=k)
    cl.fit(X_sc)
    inertia = cl.inertia_
    sil = silhouette_score(X_sc, cl.labels_)
    scores.append([k, inertia, sil])
    
score_df = pd.DataFrame(scores, columns = ['k', 'inertia', 'silhouette'])
```

```python
score_df
```

```python
axs = score_df.plot(x = 'k', subplots = True,
                   layout = (1, 2),
                   figsize = (14, 7),
                   marker = 'o')
```

```python
from scipy import spatial
def similarity(id1, id2):
    a = final[float_cols].loc[[id1]]
    b = final[float_cols].loc[[id2]]

    distance = spatial.distance.cosine(a[a.columns], b[b.columns])
    return distance
```

```python
def songs2(i):
    distances = {}
    group = final.loc[final['cluster'] == final["cluster"].loc[i]]
    float_cols = group.dtypes[group.dtypes == 'float64'].index.values
    for i in list(group[float_cols].index):
        dist = similarity(d, i)
        distances.update({i:dist})
    return distances     
```
