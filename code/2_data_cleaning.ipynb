```python
#imports
import pandas as pd
import ast
import os

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time
import numpy as np
```

```python
client_id= #client id
client_secret= #client secret

client_credentials_manager = SpotifyClientCredentials(
    client_id=client_id, 
    client_secret=client_secret
)

sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
#import all data pulled store to list
all_files =  os.listdir('../data/')
popular = [f for f in all_files if (f.lower().endswith("popularity.csv"))]
song_feat = [f for f in all_files if (f.lower().endswith("features.csv"))]
```

```python
#concat files
pop_dfs = list()
sf_dfs = list()
for filename in popular:
     dfs = pd.read_csv(f'../data/{filename}')
     pop_dfs.append(dfs)
pop = pd.concat(pop_dfs)

for filename in song_feat:
     sfdfs = pd.read_csv(f'../data/{filename}')
     sf_dfs.append(sfdfs)
feat = pd.concat(sf_dfs)
```

```python
features = ['id', 'popularity', 'name']
```

```python
#creating columns from data
dates_list = []
album = []
artist = []
for i in pop['album']:
    temp = ast.literal_eval(i)
    dates_list.append(temp.get('release_date'))
    album.append(temp.get('name'))

for i in pop['artists']:
    res = i.strip('][').split(', ')
    res[3] = res[3].replace("'name': ", "")
    artist.append(res[3].replace("'", ''))


pop['release_date']=dates_list
pop['album'] = album
pop['artist'] = artist
```

```python
pulled_artist_list = list(set(feat['id']))
new_artists_list = list(set(pop['id']))
```

```python
merged_df =  pd.merge(feat, pop, on='id', how ='left')
```

```python
features = ['release_date', 'artist', 'name', 'album', 'popularity', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms_y', 'id']
merged_df = merged_df[features]
```

```python
merged_df.shape
```

```python
merged_df.drop_duplicates(subset = ['name', 'id'], inplace = True)
merged_df
```

```python
artist_list = list(set(merged_df['artist']))
```

```python
len(artist_list)
```

```python
artists_pulled = pd.read_csv('../data/artist_genre.csv')
```

```python
artists_pulled_list = list(set(artists_pulled['Name']))
```

```python
#creating list of artists with genres already pulled
len(artists_pulled_list)
```

```python
#removing artists that have been pulled to avoid double pull
for i in artist_list:
    if i in artists_pulled_list:
        artist_list.remove(i)
```

```python
#Final list of artists
len(artist_list)
```

```python
#pulling artist genre from API
genres = {}
count = 0
for i in artist_list:
    result = sp.search(i)
    count += 1
    try:
        track = result['tracks']['items'][0]
        artist = sp.artist(track["artists"][0]["external_urls"]["spotify"])
        genres.update({i:artist["genres"]})
    except:
        result = sp.search(i[:8])
        track = result['tracks']['items'][0]
        artist = sp.artist(track["artists"][0]["external_urls"]["spotify"])
        genres.update({i:artist["genres"]})
    if count % 50 == 0:
        time.sleep(np.random.uniform(3, 5))
    if count % 500 == 0:
        print(f'{count} artist & genre')
```

```python
df = pd.DataFrame(genres.items(), columns=['Name', 'Values'])
```

```python
df_final = pd.concat([df, artists_pulled],ignore_index=True)
```

```python
df_final.to_csv('../data/artist_genre.csv', index=False)
```

```python
final = pd.merge(merged_df, df_final,  how='left', left_on='artist', right_on = 'Name')
final.drop(columns = ['Name'], inplace=True)
```

```python
final.to_csv('../data/final_merged.csv', index=False)
```
