```python
sleep_min = 2
sleep_max = 5
features = []
pop = []
request_count = 0
for i in song_id[23032:]:
    features.append(sp.audio_features(i))
    pop.append(sp.track(i))
    request_count += 1
    if request_count % 50 == 0:
        print('working on it')
        time.sleep(np.random.uniform(sleep_min, sleep_max))
    if request_count % 1000 == 0:
        time.sleep(sleep_max)
        print('taking a short break')
    if request_count % 2000 == 0:
        print(f'{int(request_count)} Songs now in Library')
    if request_count > 30000:
        print('stopping')
        break
```

```python
features_dict = []
for i in features:
    for d in i:
       features_dict.append(d)
features_dict = [i for i in features_dict if i is not None]  
```

```python
popular_df = pd.DataFrame(pop)
feat_df = pd.DataFrame(features_dict)
```

```python
popular_df.to_csv('edm8_popularity.csv')
feat_df.to_csv('edm8_track_features.csv')
```
